```python
from mpl_toolkits.mplot3d import Axes3D         # Cargo Axes3D de mpl_toolkits.mplot3d
import matplotlib. pyplot as plt
import numpy as np
import random
import math
%matplotlib inline
```

Defino una funci칩n para calcular la distancia entre 2 dountos.

```python
def dis(x,y,z):
    return pow(pow(x,2)+pow(y,2)+pow(z,2),0.5) #teorema de pitagoras
```

Construimos una caja en 2 dimenciones de tama침o 1000x1000 con una distribuci칩n de puntos aleatorios (RR)

```python
RR = open('Tabla_Aleatorios.dat',"w")
for i in range(5000):
    a, b = 0, 0
    a = random.random()*1000 #random.random s칩lo da valores entre 0 y 1 por lo que lo multiplico por 1000
    b = random.random()*1000
    RR.write("%f  \t%f \n" % (a,b))
RR.close()
RR_1 = np.loadtxt('Tabla_Aleatorios.dat') # u = RR_1[:,0] y v = RR_1[:,1]
```

```python
#Graficamos:
plt.figure(figsize=(5,5))
plt.title('Distribuci칩n aletatoria de puntos',fontsize=10)
plt.plot(RR_1[:,0],RR_1[:,1],'k,')
plt.xlabel('u',fontsize=18)
plt.ylabel('v',fontsize=18)
plt.show()
```

## 1. Crea una caja en 2d de tama침o 1000 con las siguientes distribuciones:
#### a) puntos en una malla cuadriculada de distancia (10x10 cada ret칤cula)
#### b) 1000 puntos sobre un disco de radio 300
#### c) 1000 puntos sobre un anillo con radio mayor 300 y radio menor 290
#### d) 50 anillos con 100 puntos cada uno cuyos centros est치n distribuidos aleatoriamente sobre la caja

a) Malla cuadriculada de ditancia 

```python
DD_1 = open('Tabla_Datos_malla.dat',"w")
c, d = 0, 0
for i in range(500):
    for j in range(500):
        c, d  = i*20, j*20
        DD_1.write("%f  \t%f \n" % (c,d))
DD_1.close()
DDt_1 = np.loadtxt('Tabla_Datos.dat') # u = RR_1[:,0] y v = RR_1[:,1]
```

```python
plt.figure(figsize=(5,5))
plt.title(' puntos en una malla',fontsize=10)
plt.plot(DDt_1[:,0],DDt_1[:,1],'k,')
plt.xlabel('u',fontsize=18)
plt.ylabel('v',fontsize=18)
plt.show()
```

b) 1000 puntos sobre un disco de radio 300

```python
DD_2 = open('Tabla_Datos_disco.dat',"w")
for i in range(10000):
    r = pow(random.random()*90000,0.5)
    th = random.random()*2*math.pi
    x = r*math.cos(th)
    y = r*math.sin(th)
    DD_2.write("%f  \t%f \n" % (x,y))
DD_2.close()
DDt_2 = np.loadtxt('Tabla_Datos_disco.dat') # u = RR_1[:,0] y v = RR_1[:,1]
```

```python
plt.figure(figsize=(5,5))
plt.title(' puntos en un disco',fontsize=10)
plt.plot(DDt_2[:,0],DDt_2[:,1],'k,')
plt.xlabel('u',fontsize=18)
plt.ylabel('v',fontsize=18)
plt.show()
```

c) 1000 puntos sobre un anillo con radio mayor 300 y radio menor 290

```python
DD_3 = open('Tabla_Datos_Aro.dat',"w")
for i in range(10000):
    r = pow(random.random()*90000,0.5)
    th = random.random()*2*math.pi
    x = r*math.cos(th)
    y = r*math.sin(th)
    DD_3.write("%f  \t%f \n" % (x,y))
DD_3.close()
DDt_3 = np.loadtxt('Tabla_Datos_Aro.dat') # u = RR_1[:,0] y v = RR_1[:,1]
```

```python
plt.figure(figsize=(5,5))
plt.title(' puntos en un disco',fontsize=10)
plt.plot(DDt_3[:,0],DDt_3[:,1],'k,')
plt.xlim(-300,300)
plt.ylim(-300,300)
plt.xlabel('u',fontsize=18)
plt.ylabel('v',fontsize=18)
plt.show()
```
