# the Fourier transform

```python
import matplotlib. pyplot as plt
import numpy as np
import math
from scipy.integrate import odeint
```

```python
def gaus (x,a,m):
    return np.exp(-a**2*np.power((x-m),2)/2)
```

```python
def fourier_trans(fun,x):
    N = len(fun)
    u = np.linspace(-N/2,N/2,N)
    Fun_u = []
    for i in range(N-1):
        ex = np.exp(-complex(0,1)*2*np.divide(np.pi*u[i]*x[:-1],N))
        arg = fun[:-1]*ex
        Fun_u.append(np.sum(arg.real)/N)
    return Fun_u
```

```python
x = np.linspace(-4,4,100)
f_x = gaus(x,1,0)
f_u = fourier_trans(f_x,x)
```

```python
plt.plot(x,f_x)
plt.plot(x[:-1],f_u)
```
