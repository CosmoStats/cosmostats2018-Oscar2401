###  3) Construye una caja de lado 1000 en 2 y 3 dimensiones de una distribución aletatoria de puntos.
### 4) Código que calcule histograma de distancias para 2d y 3d de una distribución de puntos aleatorios, escogiendo el tamaño de los bins, y normalizada a que su area sea 1.

```python
from mpl_toolkits.mplot3d import Axes3D         # Cargo Axes3D de mpl_toolkits.mplot3d
import matplotlib. pyplot as plt
import numpy as np
import random
import math
%matplotlib inline
```

Defino una función para calcular la distancia entre 2 dountos.

```python
def dis(x,y,z):
    return pow(pow(x,2)+pow(y,2)+pow(z,2),0.5)
```

#### Distribución en 2d

Construimos una caja en 2 dimenciones con una distribución de puntos aleatorios

```python
m, n = [], []
for i in range(5000):
    a=random.random()*1000
    b=random.random()*1000
    m.append(a)
    n.append(b)
```

```python
#Graficamos:
plt.figure()
plt.title(u'Distribución aletatoria de puntos')
plt.plot(m,n,'k,')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.show()
```

#### Histograma de distancias:

Obtenemos las distancias entre cada punto sin considerar los puntos iguales y dos distancias entre los mismos puntos.

```python
dis_2d = []
N = len(m)
for i in range(N):
    for j in range(N):
        if i != j: # No consideramos distancias entre mismos puntos.
            if i > j: #No consideramos las mismas distancias entre puntos.
                dis_2d.append(dis(m[i]-m[j],n[i]-n[j],0))
        
```

```python
histo_2d = np.array(dis_2d)
inter = 50
plt.figure()
plt.hist(histo_2d, bins=inter, color = 'k' , alpha =0.5)
plt.title("Histograma de distancias")
plt.ylabel("Frecuencia") 
plt.xlabel("Distancia")
plt.show()
```

#### Distribución en 3d

Construimos una caja en 2 dimenciones con una distribución de puntos aleatorios

```python
u,v,w=[],[],[]
for i in range(5000):
    x=random.random()*1000
    y=random.random()*1000
    z=random.random()*1000
    u.append(x)
    v.append(y)
    w.append(z)
```

```python
#Graficamos
plt.figure().gca(projection = '3d')
plt.title(u'Distribución aletatoria de puntos')
plt.plot(u,v,w,'k,')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.show()
```

#### Histograma de distancias:

Obtenemos las distancias entre cada punto sin considerar los puntos iguales y dos distancias entre los mismos puntos.

```python
dis_3d = []
N = len(m)
for i in range(N):
    for j in range(N):
        if i != j :
            if i > j:
                dis_3d.append(dis(u[i]-u[j],v[i]-v[j],w[i]-w[j]))
```

```python
histo_3d = np.array(dis_3d)
inter = 50
plt.figure()
plt.hist(histo_3d, bins=inter, color = 'k' , alpha =0.5)
plt.title("Histograma de distancias")
plt.ylabel("Frecuencia") 
plt.xlabel("Distancia")
plt.show()
```
