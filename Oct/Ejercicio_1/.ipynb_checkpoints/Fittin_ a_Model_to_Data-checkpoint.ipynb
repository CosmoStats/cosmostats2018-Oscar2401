# Prueba de ajuste de un modelo. 

```python
import numpy as np
import matplotlib. pyplot as plt
```

## El modelo generativo probabil√≠stico.

Escojemos los paramestros reales:

```python
m_true = -0.9594
b_true = 4.294
f_true = 0.534
```

Genera algunos datos sinteticos del modelo

```python
N = 50
x = np.sort(10*np.random.rand(N))
yerr = 0.1 + 0.5*np.random.rand(N)
y = m_true*x+b_true
y += np.abs(f_true*y) * np.random.randn(N)
y += yerr * np.random.randn(N)
```

```python
X0 = np.linspace(np.amin(x),np.amax(x))
Y0 = m_true*X0 + b_true
```

GRAFICA:

```python
plt.figure(1)
plt.figure(figsize=(8,8))
plt.xlabel('$x$',fontsize = 18)
plt.ylabel('$y$',fontsize = 18)
plt.errorbar(x, y, yerr=yerr, color='0' ,fmt=".")
plt.plot(X0,Y0,color='0.5')
```

```python
A = np.vstack((np.ones_like(x), x)).T
C = np.diag(yerr * yerr)
cov = np.linalg.inv(np.dot(A.T, np.linalg.solve(C, A)))
b_ls, m_ls = np.dot(cov, np.dot(A.T, np.linalg.solve(C, y)))
```

```python
print('\n Ecuaci√≥n: y =', m_ls, 'x ', b_ls)
```

```python
X_ls = np.linspace(np.amin(x),np.amax(x))
Y_ls = m_ls*X_ls + b_ls
plt.figure(figsize=(8,8))
plt.xlabel('$x$',fontsize = 18)
plt.ylabel('$y$',fontsize = 18)
plt.errorbar(x, y, yerr=yerr, color='0' ,fmt=".")
plt.plot(X0,Y0,color='0.5')
plt.plot(X_ls,Y_ls,'-.')
```

## Estimaci√≥n del maximo Likelihood 

```python
def lnlike(theta, x, y, yerr):
    m, b, lnf = theta
    model = m * x + b
    inv_sigma2 = 1.0/(yerr**2 + model**2*np.exp(2*lnf))
    return -0.5*(np.sum((y-model)**2*inv_sigma2 - np.log(inv_sigma2)))
```

## Minimizacion 

```python
def lnprior(theta):
    m, b, lnf = theta
    if -5.0 < m < 0.5 and 0.0 < b < 10.0 and -10.0 < lnf < 1.0:
        return 0.0
    return -np.inf
```

```python
def lnprob(theta, x, y, yerr):
    lp = lnprior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(theta, x, y, yerr)
```

```python
ndim, nwalkers = 3, 100
pos = [result["x"] + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]
```

```python
import emcee
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=(x, y, yerr))
```

```python
p=sampler.run_mcmc(pos, 500)
```

```python
np.array(p)
step=np.arange(100)

plt.figure(2)
plt.figure(figsize=(15,8))

plt.subplot(331)
plt.title("histogram of $H_0$")
plt.plot(step,p[0][:,0])

plt.subplot(323)
plt.plot(step,p[0][:,1])

plt.subplot(313)
plt.plot(step,p[0][:,2])
plt.xlabel("step number",fontsize=18)

plt.show()
```

```python
samples = sampler.chain[:, 50:, :].reshape((-1, ndim))
```
