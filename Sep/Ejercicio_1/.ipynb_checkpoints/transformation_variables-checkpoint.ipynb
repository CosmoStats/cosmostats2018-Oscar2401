# Uniform Distribution: Graphics

Reproduce the plots of the Uniform distribution of points inside a circle, and a sphere.

```python
import matplotlib. pyplot as plt
import numpy as np
import random
import math
%matplotlib inline
```


## Draw samples from a specific distribution. Transformation of variables.

Uniform for $r<1$

1) generate $(x,y)$ from U so
that:

\begin{equation}
r=\sqrt{x^2+y^2}<1
\end{equation}


```python
u=[]
v=[]

for i in range(10000):
    x=random.randrange(-1000000,1000000)/1000000
    y=random.randrange(-1000000,1000000)/1000000
    u.append(x)
    v.append(y)

plt.xlim(-1,1)
plt.ylim(-1,1)


for i in range(len(u)):  
    if pow((pow(u[i],2)+pow(v[i],2)),0.5) < 1:
        plt.plot(u[i],v[i],'b,') 
    else:
        plt.plot(u[i],v[i],'r,')
    
plt.show()
```

Non-Uniform for $r<1$

2) generate $r$ and
$\theta$, from U.



```python
r=[]
tetha=[]

for i in range(10000):
    x=random.randrange(-1000000,1000000)/1000000
    y=random.randrange(-1000000,1000000)/1000000
    r.append(pow((pow(x,2)+pow(y,2)),0.5))
    tetha.append(math.degrees(math.atan(y/x)))
    
plt.xlabel("x")
plt.ylabel("y")

for i in range(len(r)):  
    if r[i] < 1:
        plt.plot((r[i]*r[i])*math.cos(tetha[i]),(r[i]*r[i])*math.sin(tetha[i]),'b,') 
    else:
        plt.plot((r[i])*math.cos(tetha[i]),(r[i])*math.sin(tetha[i]),'r,')
    
plt.show()
```
