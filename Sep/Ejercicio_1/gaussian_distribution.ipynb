Use a $\mu=70$ and a $\sigma=2$.

```python
#librerias a usar 
import matplotlib. pyplot as plt
import numpy as np
import random
import scipy
import scipy.special
%matplotlib inline
```

Here I made the graphic of the normal distribution function:

\begin{equation}
P(x)=\frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
\end{equation}

with the values $\sigma=2$ and $\mu=70$

```python
x=np.arange(60,80,0.1) #llena un arreglo de 60 a 80 incrementando en 0.1
sigm=2
mu=70

p=((1/(sigm*pow(2*np.pi,0.5)))*np.exp(-(pow((x-mu),2))/(2*pow(sigm,2)))) #función de distribución.

plt.xlim(60,80)
plt.title("Probability Distribution Function")
plt.xlabel("$H_0$") 
plt.ylabel("$P(H_0)$")

plt.plot(x,p,'-')
plt.show()
```

Here I finded that the cumulative distribution function for the Gaussian Distribution is:

\begin{equation}
F(x)=\frac{1}{2} \left[ 1 + efr \left( \frac{x-\mu}{\sigma \sqrt{2}} \right) \right]
\end{equation}

where $erf(x)$ is the error function.

The grafic of the CDF is:

```python
h=np.arange(60,80,0.1)   #llena un arreglo de 0 a 80 incrementando en 1
sigm=2
mu=70

f=0.5*(1+scipy.special.erf((h-mu)/(sigm*pow(2,0.5))))
plt.xlim(60,80)
plt.title("Cumulative Distribution Function")
plt.xlabel("$H_0$") 
plt.ylabel("$F(H_0)$")
plt.plot(h,f,'-')
plt.show()
```

Choose a random number between 0 and 1, and use the CDF to assign the corresponding value of H0. Generate as many as you want, and make the histogram of H0.
The CDF invers is:

\begin{equation}
\sigma \sqrt{2}efr^{-1}(2P-1)+\mu
\end{equation}

```python
H=[]
sigm=2
mu=70

for i in range(100000):
    P=random.random()
    H.append(sigm*pow(2,0.5)*scipy.special.erfinv(2*P-1)+mu) #funcion inversa de la disribución acumulada.

inter = np.arange(60.0, 80.0, 0.1)
plt.xlim([min(H), max(H)])
plt.hist(H, bins=inter)
plt.title("histogram of $H_0$")
plt.xlabel("$H_0$") 
plt.ylabel("$Frequency$")
plt.show()
```
