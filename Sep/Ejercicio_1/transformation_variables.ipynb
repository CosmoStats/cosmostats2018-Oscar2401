# Uniform Distribution: Graphics

Reproduce the plots of the Uniform distribution of points inside a circle, and a sphere.

```python
from mpl_toolkits.mplot3d import Axes3D         # Cargo Axes3D de mpl_toolkits.mplot3d
import matplotlib. pyplot as plt
import numpy as np
import random
import math
%matplotlib inline
```


## Draw samples from a specific distribution. Transformation of variables.

### 2 Dimentions

#### Uniform for $r<1$

1) generate $(x,y)$ from U so
that:

\begin{equation}
r=\sqrt{x^2+y^2}<1
\end{equation}


```python
u=[]
v=[]

for i in range(10000):
    x=random.randrange(-1000000,1000000)/1000000 #Escoje valores entre -1 y 1 para x
    y=random.randrange(-1000000,1000000)/1000000 #Escoje valores entre -1 y 1 para y
    u.append(x)
    v.append(y)

plt.xlabel("x")
plt.ylabel("y")

for i in range(len(u)):  
    if pow((pow(u[i],2)+pow(v[i],2)),0.5) < 1:
        plt.plot(u[i],v[i],'b,') 
    else:
        plt.plot(u[i],v[i],'r,')
    
plt.show()
```

#### Non-Uniform for $r<1$

2) generate $r$ and
$\theta$, from U.

Using the coordinate change:

\begin{equation}
x=r\cos\theta \\
y=r\sin\theta \\
\end{equation}

And using their Jacobian:
\begin{equation}
r
\end{equation}

```python
r=[]
tetha=[]

for i in range(10000):
    x=random.randrange(-1000000,1000000)/1000000
    y=random.randrange(-1000000,1000000)/1000000
    r.append(pow((pow(x,2)+pow(y,2)),0.5))
    tetha.append(math.degrees(math.atan(y/x)))
    
plt.xlabel("x")
plt.ylabel("y")

for i in range(len(r)):  
    rad=(r[i]*math.cos(tetha[i]))
    th=(r[i]*math.sin(tetha[i]))
    if r[i] < 1:
        plt.plot((rad*r[i]),(th*r[i]),'b,')  #Cambio de coordenada multiplicado por el jacobiano r 
    else:
        plt.plot(rad,th,'r,')
    
plt.show()
```

### 3 dimention

Graphic of uniform distribution with:
   
\begin{equation}
    r=\sqrt{x^2+y^2+z^2}< 1
\end{equation}

```python
u=[]
v=[]
w=[]

for i in range(1000):
    x=random.randrange(-1000000,1000000)/1000000
    y=random.randrange(-1000000,1000000)/1000000
    z=random.randrange(-1000000,1000000)/1000000
    u.append(x)
    v.append(y)
    w.append(z)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(len(u)):  
    if pow((pow(u[i],2)+pow(v[i],2)+pow(w[i],2)),0.5) < 1:
        ax.scatter(u[i],v[i],w[i], c='r', marker='.') 
        
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
        
plt.show()
    
```

3) generate $x$,$y$ from U generate $r$, $\theta$, and $\phi$ 

using the coordinate change:

\begin{equation}
x=r\sin\theta\cos\phi \\
y=r\sin\theta\sin\phi \\
z=rcos\theta
\end{equation}

Using their Jacobian:
\begin{equation}
r^{2}\sin\theta
\end{equation}

generate $r$, $\theta$ and $\phi$ from U.

```python


R=[] #Radio
T=[] #Theta
P=[] #Phi

for i in range(1000):
    x=random.randrange(-1000000,1000000)/1000000
    y=random.randrange(-1000000,1000000)/1000000
    z=random.randrange(-1000000,1000000)/1000000
    R.append(pow((pow(x,2)+pow(y,2)+pow(z,2)),0.5))
    T.append(math.degrees(math.atan(pow((pow(x,2)+pow(y,2)),0.5)/z)))
    P.append(math.degrees(math.atan(y/x)))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(len(R)):
    r=((R[i])*math.sin(T[i])*math.cos(P[i]))*(R[i]**2)*math.sin(T[i])  #Cmbio de coordenadas multiplicando por el jacobiano. 
    t=((R[i])*(math.sin(T[i])*math.sin(P[i])))*(R[i]**2)*math.sin(T[i])
    p=((R[i])*math.cos(T[i]))*(R[i]**2)*math.sin(T[i])
    if R[i] < 1:
        ax.scatter(r,t,p, c='r', marker='.') 
        
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
        
plt.show()
```
